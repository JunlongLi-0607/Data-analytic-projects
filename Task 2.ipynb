Calculate_similarity method

```python
# this method is one of the way of comparing similarity between two stores(pearson value)
def calculate_similarity(df, trial_store_nbr, control_store_nbr, metric_cols, pre_trial_months):
    """
    df: store_metrics）
    trial_store_nbr: 
    control_store_nbr: 
    metric_cols: the name of metrics that need to be compared，such as ['total_sales', 'n_customers']
    pre_trial_months: such as ['201807', ..., '201901']
    
    """
    # Filtering the data of pre-trial of trial_store_nbr and control_store_nbr
    trial_data = df[(df['STORE_NBR'] == trial_store_nbr) & (df['month_id'].isin(pre_trial_months))]
    control_data = df[(df['STORE_NBR'] == control_store_nbr) & (df['month_id'].isin(pre_trial_months))]
    
    # Sorting the data base on months 
    trial_data = trial_data.sort_values(by='month_id').reset_index(drop=True)
    control_data = control_data.sort_values(by='month_id').reset_index(drop=True)
    
    # if the length of months are unmatched, directly returm NaN
    if len(trial_data) != len(pre_trial_months) or len(control_data) != len(pre_trial_months):
        return None

    correlations = []
    for col in metric_cols:
        corr = trial_data[col].corr(control_data[col])
        correlations.append(corr)
    
    # Return：the pearson values of multiple metrics
    return sum(correlations) / len(correlations)
```

```python
trial_stores = [77, 86, 88]
pre_trial_months = ['2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12', '2019-01']
metrics_to_compare = ['total_sales', 'n_customers', 'transactions_per_customer']

pre_trial_data = store_month_metrics[store_month_metrics['month_id'].isin(pre_trial_months)]# Filtering all data whose month_id is in pre_trial_months

#Calculating how many months for each store
store_month_counts = (
    pre_trial_data.groupby('STORE_NBR')['month_id']
    .nunique()
    .reset_index()
    .rename(columns={'month_id': 'month_count'})
)

#Filtering those store(valid_control_candidates) which has at least the data of 6 months out of 7 in the pre_trial period
valid_control_candidates = store_month_counts[store_month_counts['month_count'] >= 6]
#Filtering those stores(valid_control_candidates) which are not the trial stores
valid_control_candidates = valid_control_candidates[~valid_control_candidates['STORE_NBR'].isin(trial_stores)]

candidate_control_stores = valid_control_candidates['STORE_NBR'].tolist()
```

```python
best_matches = {}

for trial_store in trial_stores:
    best_score = -1
    best_match = None

    for control_store in candidate_control_stores:
        score = calculate_similarity(store_month_metrics, trial_store, control_store, metrics_to_compare, pre_trial_months)
        if score is not None and score > best_score:
            best_score = score
            best_match = control_store

    best_matches[trial_store] = {'control_store': best_match, 'similarity_score': best_score}


for trial, result in best_matches.items():
    print(f"Trial Store {trial} matched with Control Store {result['control_store']} (Similarity: {result['similarity_score']:.3f})")
```

Calculate_magnitude_distance_scores method

```python
# this method is the other way of comparing the similarity between two stores(direct distance)
def calculate_magnitude_distance_scores(df, trial_store_nbr, candidate_stores, metric_cols, pre_trial_months):
    """
    df: store_metrics）
    trial_store_nbr: 
    control_store_nbr: 
    metric_cols: the name of metrics that need to be compared，such as ['total_sales', 'n_customers']
    pre_trial_months: such as ['201807', ..., '201901']
    """
    trial_data = df[(df['STORE_NBR'] == trial_store_nbr) & (df['month_id'].isin(pre_trial_months))]
    trial_data = trial_data.sort_values(by='month_id').reset_index(drop=True)

    scores = []
    
    for control_store in candidate_stores:
        control_data = df[(df['STORE_NBR'] == control_store) & (df['month_id'].isin(pre_trial_months))]
        control_data = control_data.sort_values(by='month_id').reset_index(drop=True)

        # make sure the data is in align
        if len(trial_data) != len(pre_trial_months) or len(control_data) != len(pre_trial_months):
            continue
        
        # Calculating the total distance (多个指标的绝对差值和)
        total_distance = 0
        for col in metric_cols:
            distance = abs(trial_data[col] - control_data[col]).sum()
            total_distance += distance

        scores.append({'control_store': control_store, 'distance': total_distance})
    
    scores_df = pd.DataFrame(scores)

    if scores_df.empty:
        return pd.DataFrame(columns=['control_store', 'distance', 'similarity_score'])

    # 归一化成分数（越接近1越相似）
    min_dist = scores_df['distance'].min()
    max_dist = scores_df['distance'].max()
    scores_df['similarity_score'] = 1 - (scores_df['distance'] - min_dist) / (max_dist - min_dist + 1e-9)  
    return scores_df.sort_values('similarity_score', ascending=False)
```

```python
best_matches = {}

for trial_store in trial_stores:
    scores_df = calculate_magnitude_distance_scores(
        df=store_month_metrics,
        trial_store_nbr=trial_store,
        candidate_stores=candidate_control_stores,
        metric_cols=metrics_to_compare,
        pre_trial_months=pre_trial_months
    )

    if not scores_df.empty:
        top_match = scores_df.iloc[0]
        best_matches[trial_store] = {
            'control_store': int(top_match['control_store']),
            'similarity_score': top_match['similarity_score']
        }
    else:
        best_matches[trial_store] = {
            'control_store': None,
            'similarity_score': None
        }

# -----------------------------
# Print the match outcome
# -----------------------------
for trial, result in best_matches.items():
    if result['control_store'] is not None:
        print(f"Trial Store {trial} matched with Control Store {result['control_store']} (Similarity Score: {result['similarity_score']:.3f})")
    else:
        print(f"Trial Store {trial} has no valid control match.")
```

```python
# Assume that the trial store you define is:：
trial_store = 77
pre_trial_months = ['2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12', '2019-01']
candidate_stores = candidate_control_stores
corr_weight = 0.5

# ========== 1. Calculating correlation ==========
sales_corr = []
cust_corr = []

for control in candidate_stores:
    sales_c = calculate_similarity(store_month_metrics, trial_store, control, ['total_sales'], pre_trial_months)
    cust_c = calculate_similarity(store_month_metrics, trial_store, control, ['n_customers'], pre_trial_months)
    sales_corr.append({'control_store': control, 'sales_corr': sales_c})
    cust_corr.append({'control_store': control, 'cust_corr': cust_c})

sales_corr_df = pd.DataFrame(sales_corr)
cust_corr_df = pd.DataFrame(cust_corr)

# ========== 2. Calculating magnitude distance ==========
sales_mag_df = calculate_magnitude_distance_scores(
    df=store_month_metrics,
    trial_store_nbr=trial_store,
    candidate_stores=candidate_stores,
    metric_cols=['total_sales'],
    pre_trial_months=pre_trial_months
).rename(columns={'similarity_score': 'sales_mag'})

cust_mag_df = calculate_magnitude_distance_scores(
    df=store_month_metrics,
    trial_store_nbr=trial_store,
    candidate_stores=candidate_stores,
    metric_cols=['n_customers'],
    pre_trial_months=pre_trial_months
).rename(columns={'similarity_score': 'cust_mag'})

# ========== 3. Combine all four score tables ==========
merged = sales_corr_df.merge(sales_mag_df, on='control_store')
merged = merged.merge(cust_corr_df, on='control_store')
merged = merged.merge(cust_mag_df, on='control_store')

# ========== 4. Cuculating the overall score ==========
merged['score_nSales'] = corr_weight * merged['sales_corr'] + (1 - corr_weight) * merged['sales_mag']
merged['score_nCustomers'] = corr_weight * merged['cust_corr'] + (1 - corr_weight) * merged['cust_mag']

# ========== 5. Final composite score（two metrics averagely） ==========
merged['composite_score'] = (merged['score_nSales'] + merged['score_nCustomers']) / 2

# ========== 6. Print the final outcome ==========
merged_sorted = merged.sort_values(by='composite_score', ascending=False)
print(merged_sorted[['control_store', 'sales_corr', 'sales_mag', 'cust_corr', 'cust_mag', 'composite_score']].head())
```

```python
trial_store = 77
control_store = 233
pre_trial_months = ['2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12', '2019-01']
trial_period_months = ['2019-02', '2019-03', '2019-04']
scaling_factor = 1.024  # Replace with the scaling factor calculated

# --- STEP 1: Calculate pre-trial uplift percentage differences ---
pre_trial_data = store_month_metrics[
    (store_month_metrics['STORE_NBR'].isin([trial_store, control_store])) &
    (store_month_metrics['month_id'].isin(pre_trial_months))
].copy()

# Adding the sale of column after scaling (for control store only)
pre_trial_data['scaled_sales'] = pre_trial_data.apply(
    lambda row: row['total_sales'] * scaling_factor if row['STORE_NBR'] == control_store else row['total_sales'],
    axis=1
)

# Stores seperated
trial_sales_pre = pre_trial_data[pre_trial_data['STORE_NBR'] == trial_store][['month_id', 'scaled_sales']].rename(columns={'scaled_sales': 'trial_sales'})
control_sales_pre = pre_trial_data[pre_trial_data['STORE_NBR'] == control_store][['month_id', 'scaled_sales']].rename(columns={'scaled_sales': 'control_sales'})

# Combine and caculate uplift
pre_uplift_df = pd.merge(trial_sales_pre, control_sales_pre, on='month_id')
pre_uplift_df['percentage_diff'] = (
    (pre_uplift_df['trial_sales'] - pre_uplift_df['control_sales']) / pre_uplift_df['control_sales']
) * 100

# --- STEP 2: Calculate standard deviation and freedom ---
std_dev = np.std(pre_uplift_df['percentage_diff'], ddof=1)  # standard deviation
degrees_freedom = len(pre_trial_months) - 1  # usually n-1
print(f"Standard Deviation of Pre-Trial % Diff: {std_dev:.2f}, Degrees of Freedom: {degrees_freedom}")

# --- STEP 3: Calculating the uplift and t value in trial period ---
trial_data = store_month_metrics[
    store_month_metrics['STORE_NBR'].isin([trial_store, control_store]) &
    store_month_metrics['month_id'].isin(trial_period_months)
].copy()

trial_data['scaled_sales'] = trial_data.apply(
    lambda row: row['total_sales'] * scaling_factor if row['STORE_NBR'] == control_store else row['total_sales'],
    axis=1
)

trial_sales = trial_data[trial_data['STORE_NBR'] == trial_store][['month_id', 'scaled_sales']].rename(columns={'scaled_sales': 'trial_sales'})
control_sales = trial_data[trial_data['STORE_NBR'] == control_store][['month_id', 'scaled_sales']].rename(columns={'scaled_sales': 'control_sales'})

uplift_trial_df = pd.merge(trial_sales, control_sales, on='month_id')
uplift_trial_df['percentage_diff'] = (
    (uplift_trial_df['trial_sales'] - uplift_trial_df['control_sales']) / uplift_trial_df['control_sales']
) * 100

# --- STEP 4: Calculating t value and see its significance  ---
uplift_trial_df['t_value'] = uplift_trial_df['percentage_diff'] / std_dev
critical_t = t.ppf(0.975, df=degrees_freedom)
uplift_trial_df['significant'] = uplift_trial_df['t_value'].abs() > critical_t

# --- STEP 5: Outcome printed ---
print("\nT-Test Results:")
print(uplift_trial_df[['month_id', 'percentage_diff', 't_value', 'significant']])
```
